```python
import pandas as pd
import numpy as np
import h2o
```

```python
from h2o.estimators import H2OGeneralizedLinearEstimator
```

```python
train = pd.read_csv("D:/ML/Kaggle/playground-series-s5e3/train.csv")
test = pd.read_csv("D:/ML/Kaggle/playground-series-s5e3/test.csv")
```

```python
train["rainfall"] = train["rainfall"].astype('category')
```

```python
h2o.init()
```

```python
train_h2o = h2o.H2OFrame(train)
test_h2o = h2o.H2OFrame(test)
```

```python
train, valid = train_h2o.split_frame(ratios=[.7], seed=1234)
```

```python
y = "rainfall"
```

```python
x = [col for col in train_h2o.columns if col not in ["id", "rainfall"]]
print(x)
```

```python
model = H2OGeneralizedLinearEstimator(
    family = "binomial", 
    lambda_ = 0, 
    compute_p_values = True,
    generate_scoring_history = True 
)
```

```python
model.train(x, y, training_frame=train)
```

```python
perf = model.model_performance(valid)
```

```python
pred = model.predict(test_data=test_h2o)
```

```python
pred_df = pred.as_data_frame()
```

```python
pred_df['id'] = test['id']
```

```python
pred_df = pred_df[["id", "p1"]]
```

```python
pred_df.to_csv('submission.csv', index=False)
```
